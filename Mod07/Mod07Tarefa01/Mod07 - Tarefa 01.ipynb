    Dica: 
        1. inicie uma lista vazia

3. Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
df_metadados = pd.DataFrame(data=df.dtypes.reset_index())
df_metadados.columns = ['colunas', 'dtypes']
df_metadados
````

````python
qtd_categorias = list()
for coluna in df.columns:
    qtd_categorias.append(df[coluna].nunique())
df_metadados['qtd_categorias'] = qtd_categorias
df_metadados
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
colunas_alvo = ['sexo', 'posse_de_veiculo', 'posse_de_imovel']
df_dummies = pd.get_dummies(data=df.copy(), 
                             columns=colunas_alvo, 
                             drop_first=True)
df_dummies.info()
````

````python
df_dummies['mau'] = df_dummies['mau'].astype('int64')
df_dummies['sexo_M'] = df_dummies['sexo_M'].astype('int64')
df_dummies['posse_de_veiculo_Y'] = df_dummies['posse_de_veiculo_Y'].astype('int64')
df_dummies['posse_de_imovel_Y'] = df_dummies['posse_de_imovel_Y'].astype('int64')
````

````python
df_dummies.info()
````

````python
df_dummies.head(5)
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
df_dummies['mau'].value_counts()
````

````python
cruzado_veiculo = pd.crosstab(df_dummies['posse_de_veiculo_Y'], df_dummies['mau'])
cruzado_veiculo
````

````python
cruzado_veiculo_pct = cruzado_veiculo.div(cruzado_veiculo.sum(axis = 1), axis = 0)*100
cruzado_veiculo_pct 
````

````python
cruzado_email = pd.crosstab(df_dummies['possui_email'], df_dummies['mau'])
cruzado_email 
````

````python
cruzado_email_pct = cruzado_email.div(cruzado_email.sum(axis = 1), axis = 0)*100
cruzado_email_pct 
````

Acredito que ambas as variáveis possuem o mesmo peso, pois ambas têm cerca de 2% de influência
