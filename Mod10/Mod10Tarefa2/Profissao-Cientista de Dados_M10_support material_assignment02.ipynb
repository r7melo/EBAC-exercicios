````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
df.info()
````

````python
ax_imovel = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df, dodge=True, ci = 90)
````

### Insight:
 - De acordo com o gráfico acima, quem possui imóvel tem a renda maior.

````python
ax_veiculo = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df, dodge=True, ci = 90)
````

### Insight:

 - De acordo com o gráfico acima, quem possui veículo tem a renda maior.

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Gráfico 1: Posse de Imóvel
ax_imovel_mau = sns.pointplot(x="posse_de_imovel", y="renda", hue='mau',
                              data=df, dodge=True, ci=90, ax=ax1)
ax_imovel_mau.set_title('Posse de Imóvel x Renda X inadimplência')
ax_imovel_mau.set_xlabel('Posse de Imóvel')
ax_imovel_mau.set_ylabel('Renda')
ax_imovel_mau.legend(title='Inadimplente')

# Gráfico 2: Posse de Veículo
ax_veiculo_mau = sns.pointplot(x="posse_de_veiculo", y="renda", hue='mau',
                               data=df, dodge=True, ci=90, ax=ax2)
ax_veiculo_mau.set_title('Posse de Veículo x Renda X inadimplência')
ax_veiculo_mau.set_xlabel('Posse de Veículo')
ax_veiculo_mau.set_ylabel('Renda')
ax_veiculo_mau.legend(title='Inadimplente')

plt.tight_layout()
plt.show()
````

### Insights:
 - De acordo com os gráficos acima, podemos observar que a maioria dos inadimplentes não possui imóvel, e aqueles que possuem têm uma renda média de 4500 reais.
 - Por outro lado, a maioria dos inadimplentes que possuem veículo tem uma renda média de 4300 reais, enquanto os bons pagadores têm uma renda superior a 5000 reais.
