####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui
def pvl_estado_municipio():
    lista_url = list()
    ufs_brasileiras = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']
    while True:
        estado_uf = input('Digite a UF do estado:[Digite 999 para parar] ').strip().upper()
        if estado_uf not in ufs_brasileiras and estado_uf != '999':
            print('Resposta incorreta, tente novamente!')
        elif estado_uf == '999':
            break
        else:
            while True:
                estado_municipio = input('Você gostaria da URL do estado ou município? ').strip().upper()[0]
                if estado_municipio == 'E':
                    url_e = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={estado_uf}&tipo_interessado=Estado'
                    lista_url.append(url_e)
                    print(f'URL para o PVL da UF do estado informado: \n{url_e}')
                    break
                    lista_url.clear()
                elif estado_municipio == 'M':
                    url_m = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={estado_uf}&tipo_interessado=Munic%C3%ADpio'
                    lista_url.append(url_m)
                    print(f'URL para o PVL de município informado pela UF:\n{url_m}')
                    break
                    lista_url.clear()
                elif 'ME' in estado_municipio:
                    print('Resposta incorreta, tente novamente!')
            break
        
    return lista_url
````

````python
# 2) Seu código aqui
url = pvl_estado_municipio()
url = '\n'.join(url)
r_url = requests.get(url)
data_json_url = r_url.json()
df_url = pd.DataFrame(data_json_url['items'])
df_url.head(3)
````

````python
# 2) Seu código aqui
df_url['status'].value_counts()
````

````python
# 2) Seu código aqui
arquivado_decurso_mg = df_url['status'].value_counts().get('Arquivado por decurso de prazo', 0)
print(f'{arquivado_decurso_mg} solicitação foi consultada com status em "Arquivado por decurso de prazo" para MG.')
````

````python
#Qual é o município da Bahia com mais solicitações deferidas?
# 3) Seu código aqui
url = pvl_estado_municipio()
url = '\n'.join(url)
r_url = requests.get(url)
data_json_url = r_url.json()
df_url = pd.DataFrame(data_json_url['items'])
df_url.head(3)
````

````python
# 3) Seu código aqui
solic_defiridas_ba= df_url['interessado'].loc[df_url.status == 'Deferido']
solic_defiridas_ba.value_counts()
````

````python
#3) Qual é o município da Bahia com mais solicitações deferidas?
print('O munícipio com mais solicitaçãoes deveridas foi Lúis Eduardo Magalhães com 16.')
````
