#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
%matplotlib widget

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

font_options = {
    'family':'monospace',
    'weight':'bold',
    'size':10
}
plt.rc('font', **font_options)
```

```python
sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))

sinasc_ = sinasc[[ 'APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]

data = {}

for col in sinasc_.columns:
    data[col] = [len(sinasc_[col].dropna())]

df = pd.DataFrame(data)

df.plot.bar()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 10))
ac = sinasc[['APGAR5','CONSULTAS']]

sns.barplot(x='APGAR5', y='CONSULTAS', data=ac, errorbar=None)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 20))

fig, axes = plt.subplots(3,2)
fig.subplots_adjust(hspace=.3)

    
sns.countplot(ax=axes[0,0], x='APGAR1', data=sinasc)
sns.countplot(ax=axes[0,1], x='APGAR5', data=sinasc)
sns.countplot(ax=axes[1,0], x='SEXO', data=sinasc)
sns.countplot(ax=axes[1,1], x='GESTACAO', data=sinasc)
sns.countplot(ax=axes[2,0], x='ESTCIVMAE', data=sinasc)

axes[1,1].tick_params(axis='x', rotation=30)
    
```
