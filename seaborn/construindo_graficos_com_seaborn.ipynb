```python
%matplotlib widget

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

font_options = {
    'family':'monospace',
    'weight':'bold',
    'size':10
}
plt.rc('font', **font_options)
```

```python
sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(data=sinasc, x='APGAR1')
plt.title('Distribuição de APGAR1')
```

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(data=sinasc, x='APGAR5')
plt.title('Distribuição de APGAR5')
```

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(data=sinasc, y='SEXO')
plt.title('Distribuição de SEXO')
```

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(data=sinasc, y='GESTACAO')
plt.title('Distribuição de GESTACAO')
```

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(data=sinasc, x='ESTCIVMAE')
plt.title('Distribuição de ESTCIVMAE')
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))

mediac = sinasc.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

sns.barplot(data=mediac, x='APGAR5', y='CONSULTAS', errorbar=None)
plt.title('Média de APGAR5 por CONSULTAS')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 20))

fig, axes = plt.subplots(3,2)
fig.subplots_adjust(hspace=.3)

    
sns.countplot(ax=axes[0,0], x='APGAR1', data=sinasc)
sns.countplot(ax=axes[0,1], x='APGAR5', data=sinasc)
sns.countplot(ax=axes[1,0], x='SEXO', data=sinasc)
sns.countplot(ax=axes[1,1], x='GESTACAO', data=sinasc)
sns.countplot(ax=axes[2,0], x='ESTCIVMAE', data=sinasc)

axes[1,1].tick_params(axis='x', rotation=30)
    
```
