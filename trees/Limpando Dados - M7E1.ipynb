#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df.dtypes
````

````python
df_dummies = pd.get_dummies(df, columns=['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'], drop_first=True)
df_dummies 
````

````python
df_dummies.dtypes
````

````python
df_dummies.sexo = df_dummies.sexo.map({'M':1,'F':0})
df_dummies.posse_de_veiculo = df_dummies.posse_de_veiculo.map({'Y':1,'N':0})
df_dummies.posse_de_imovel = df_dummies.posse_de_imovel.map({'Y':1,'N':0})

df_dummies
````

````python
df_dummies.dtypes
````

````python
col_bool = df_dummies.select_dtypes(include=['bool']).columns
df_dummies[col_bool] = df_dummies[col_bool].astype(int)
df_dummies
````

````python
df_dummies.dtypes
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
prev_mau_1 = pd.crosstab(df_dummies['possui_email'], df_dummies['mau'])
prev_mau_1_pct = (prev_mau_1 / prev_mau_1.values.sum()) * 100
prev_mau_1_pct
````

````python
prev_mau_2 = pd.crosstab(df_dummies['posse_de_veiculo'], df_dummies['mau'])
prev_mau_2_pct = (prev_mau_2 / prev_mau_2.values.sum()) * 100
prev_mau_2_pct
````

O 'posse de veiculo' tem mais influencia em mau, com 0.85%, do que o 'possui email', com somente 0.1
