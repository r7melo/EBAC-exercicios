```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
df = pd.read_csv('sinasc_RO_2019.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
plt.rc('figure', figsize=(8,6))
df['APGAR1'].value_counts().sort_index().plot.bar()
plt.title(label='APGAR1')
plt.tick_params(axis='x', rotation=False)
plt.tight_layout()
plt.show()
```

```python
plt.close('all')
plt.rc('figure', figsize=(8,6))
df['APGAR5'].value_counts().sort_index().plot.bar()
plt.title(label='APGAR5')
plt.tick_params(axis='x', rotation=False)
plt.show()
```

```python
plt.close('all')
plt.rc('figure', figsize=(8,6))
df['SEXO'].value_counts().sort_index().plot.bar()
plt.title(label='SEXO')
plt.tick_params(axis='x', rotation=False)
plt.tight_layout()
plt.show()
```

```python
plt.close('all')
plt.rc('figure', figsize=(8,6))
df['GESTACAO'].value_counts().sort_index().plot.bar()
plt.title(label='GESTAÇÃO')
plt.xlabel(xlabel='SEMANAS DE GESTAÇÃO')
plt.ylabel(ylabel='Total')
plt.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.close('all')
#Mudar os números pela descrição do estado cívil:
df['ESTCIVMAE'] = df['ESTCIVMAE'].astype(str)
estado_civil = {'1.0':'Solteira', '2.0':'Casada', '3.0':'Viúva', '4.0':'Divorciado', '5.0':'União consensual', '9.0':'Ignorado'}
df['ESTCIVMAE'] = df['ESTCIVMAE'].replace(estado_civil)
#gráfico
plt.rc('figure', figsize=(8,6))
df['ESTCIVMAE'].value_counts().sort_index().plot.bar()
plt.title(label='ESTADO CÍVIL DAS GESTANTES')
plt.xlabel(xlabel= 'Estado cívil')
plt.ylabel(ylabel='Total')
plt.tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.rc('figure', figsize=(10, 20))
fig, axes = plt.subplots(3, 2)
sns.countplot(ax=axes[0,0], x = 'APGAR1', data=df)
sns.countplot(ax=axes[0,1], x = 'APGAR5', data=df)
sns.countplot(ax=axes[1,0], x = 'SEXO', data=df)
sns.countplot(ax=axes[1,1], x = 'GESTACAO', data=df)
axes[1, 1].tick_params(axis='x', rotation=90)
sns.countplot(ax=axes[2,0], x = 'ESTCIVMAE', data=df)
axes[2,0].tick_params(axis='x', rotation=90)
axes[0, 0].set_ylabel('Total')
axes[0, 1].set_ylabel('Total')
axes[1, 0].set_ylabel('Total')
axes[1, 1].set_ylabel('Total')
axes[2, 0].set_ylabel('Total')
axes[2, 0].set_xlabel('Estado cívil das gestantes')
plt.tight_layout()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10,6))
sns.barplot(x='APGAR5', y='CONSULTAS', data=df)
plt.tight_layout()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
#Ajeitando o gráfico pra ficar mais fácil fácil de compreender o gráfico 'Consulta'
df['CONSULTAS'] = df['CONSULTAS'].astype(str)
qtd_consultas = {'1':'Nenhuma', '2':'De 1 a 3 ', '3': 'De 4 a 6', '4': '7 ou mais', '9':'Ignorado'}
df['CONSULTAS'] = df['CONSULTAS'].replace(qtd_consultas)
df['CONSULTAS']
#código gráfico
plt.rc('figure', figsize=(10, 10))
fig, axes = plt.subplots(2, 2)
sns.countplot(ax=axes[0,0], x = 'SEXO', data=df)
sns.countplot(ax=axes[0,1], x = 'GESTACAO', data=df)
axes[0, 1].tick_params(axis='x', rotation=90)
sns.countplot(ax=axes[1,0], x = 'ESTCIVMAE', data=df)
axes[1,0].tick_params(axis='x', rotation=90)
sns.countplot(ax=axes[1,1], x = 'CONSULTAS', data=df)
axes[1, 0].set_xlabel('Estado cívil das gestantes')
plt.tight_layout()
```
