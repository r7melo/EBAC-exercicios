#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('sinasc_RO_2019.csv')
df
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[(df['APGAR5'] >= 0) & (df['APGAR5'] <= 3), 'APGAR5_CAT'] = 'asfixia grave'
df.loc[(df['APGAR5'] >= 4) & (df['APGAR5'] <= 5), 'APGAR5_CAT'] = 'asfixia moderada'
df.loc[(df['APGAR5'] >= 6) & (df['APGAR5'] <= 7), 'APGAR5_CAT'] = 'asfixia leve'
df.loc[(df['APGAR5'] >= 8) & (df['APGAR5'] <=10), 'APGAR5_CAT'] = 'normal'
df['APGAR5_CAT'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.barplot(x='GESTACAO', y='APGAR5', hue='SEXO', data=df)
plt.tick_params(axis='x', rotation=45)
plt.title('Distribuição APGAR5 X Tempo de gestação')
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabe_cruz = pd.crosstab(df['GESTACAO'],df['APGAR5_CAT'])
tabe_cruz
```

####  5) Da tabela cruzada do item anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tabe_cruz_pct = tabe_cruz.div(tabe_cruz.sum(axis = 1), axis = 0)*100
tabe_cruz_pct 
```

```python
tabe_cruz_pct.sum(axis=1)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
tabe_cruz_pct.plot(kind='bar').legend(loc='upper left', frameon = False)
plt.tick_params(axis='x', rotation=45)
plt.ylabel('Porcentagem (%)')
plt.xlabel('Duração da gestação')
plt.title('Duração da gestação X Apgar 5')
```

####  Com base no gráfico acima, chegamos a conclusão que com menos semanas de gestação os indices de asfixia sobem. Sendo assim, corrorando com a afirmação dos médicos.
