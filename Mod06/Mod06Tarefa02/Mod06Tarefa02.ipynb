```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('sinasc_RO_2019.csv')
df
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
ay = plt.gca()
box_idade = df[['IDADEMAE']].boxplot()
box_idade.set_title(label='Idade da Mãe')
ay.set_ylabel(ylabel='Idade em anos')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
ay = plt.gca()
box_peso = sns.boxplot(y='PESO', data=df)
box_peso.set_title(label='Peso dos bebês')
ay.set_ylabel(ylabel='Peso em gramas')
plt.show
```

####  3) Faça um histograma da idade da mãe

```python
ax = plt.gca()
hist_idade = df['IDADEMAE'].plot.hist()
hist_idade.set_title(label='Idade da mãe')
ax.set_xlabel(xlabel='Idade em anos')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
hist_peso = sns.displot(data=df['PESO'],  bins = 20)
hist_peso.set(title='Peso dos bebês')
hist_peso.set(xlabel = 'Peso em gramas')
plt.show()
```
